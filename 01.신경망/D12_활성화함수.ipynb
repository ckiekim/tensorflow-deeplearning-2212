#2. 활성화 함수(Activation function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

- heaviside step function

```python
def heaviside(x):
    y = np.zeros(x.size)
    y[x > 0] = 1        # x <= 0 이면 0, x > 0 이면 1
    return y
```

```python
xs = np.linspace(-5, 5, 501)
ys = heaviside(xs)
plt.plot(xs, ys)
plt.grid()
plt.title('Heaviside step function')
plt.show()
```

- sigmoid 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sigmoid(np.array([-10, -5, 0, 5, 10]))
```

```python
xs = np.linspace(-5, 5, 501)
ys = sigmoid(xs)
plt.plot(xs, ys, label='sigmoid')
plt.plot(xs, ys*(1-ys), label='derivative')
plt.grid()
plt.legend()
plt.title('Sigmoid function')
plt.show()
```

- Hyperbolic tangent function

```python
ys = np.tanh(xs)
plt.plot(xs, ys)
plt.grid()
plt.title('Hyperbolic tangent function')
plt.show()
```

- Relu function

```python
def relu(x):
    return np.maximum(0, x)
```

```python
relu(np.array([-10, -5, 0, 5, 10]))
```

```python
ys = relu(xs)
plt.plot(xs, ys)
plt.grid()
plt.title('ReLU function')
plt.show()
```

- Leaky ReLU function

```python
def leaky_relu(x):
    return np.maximum(0.01 * x, x)
```

```python
ys = leaky_relu(xs)
plt.plot(xs, ys)
plt.grid()
plt.title('Leaky ReLU function')
plt.show()
```

- Softplus function

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
ys = softplus(xs)
plt.plot(xs, ys)
plt.grid()
plt.title('Softplus function')
plt.show()
```

- 종합

```python
plt.figure(figsize=(15,7))
plt.subplot(231)            # 행이 2개, 열이 3개인 그래프의 첫번째
plt.plot(xs, heaviside(xs)), plt.grid(), plt.title('step function')
plt.subplot(232)
plt.plot(xs, sigmoid(xs)), plt.grid(), plt.title('sigmoid function')
plt.subplot(233)
plt.plot(xs, np.tanh(xs)), plt.grid(), plt.title('hyperbolic tangent function')
plt.subplot(234)
plt.plot(xs, relu(xs)), plt.grid(), plt.title('relu function')
plt.subplot(235)
plt.plot(xs, leaky_relu(xs)), plt.grid(), plt.title('leaky relu function')
plt.subplot(236)
plt.plot(xs, softplus(xs)), plt.grid(), plt.title('softplus function')
plt.show()
```
