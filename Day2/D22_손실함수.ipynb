# 손실 함수(Loss function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yhat1 = np.array([0.1, 0.8, 0.1, 0])
yhat2 = np.array([0.1, 0.2, 0.7, 0])
y = np.array([0, 1, 0, 0])
N = y.size
```

- 평균 제곱 오차(MSE: Mean Squared Error)

```python
mse1 = np.sum(np.square(yhat1 - y)) / N
mse2 = np.sum(np.square(yhat2 - y)) / N
mse1, mse2
```

- 평균 제곱근 오차(RMSE: Root Mean Squared Error)

```python
rmse1 = np.sqrt(mse1)
rmse2 = np.sqrt(mse2)
rmse1, rmse2
```

- 평균 절대 오차(MAE: Mean Absolute Error)

```python
mae1 = np.sum(np.abs(yhat1 - y)) / N
mae2 = np.sum(np.abs(yhat2 - y)) / N
mae1, mae2
```

- Categorical Cross Entropy

```python
def cross_entropy_error(y, t):
    if y.ndim == 1:
        y = y.reshape(1, y.size)
        t = t.reshape(1, t.size)
    if y.size == t.size:
        t = t.argmax(axis=1)
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), t])) / batch_size
```

```python
cce1 = cross_entropy_error(yhat1, y)
cce2 = cross_entropy_error(yhat2, y)
cce1, cce2
```

- Binary Cross Entropy

```python
xs = np.linspace(0.001, 0.999, 999)
plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.plot(xs, np.log(xs)), plt.title('log x'), plt.grid()
plt.subplot(222)
plt.plot(xs, -np.log(xs)), plt.title('-log x'), plt.grid()          # x축 대칭
plt.subplot(223)
plt.plot(xs, np.log(1 - xs)), plt.title('log (-(x-1))'), plt.grid() # x를 1만큼 평행이동후 y축 대칭
plt.subplot(224)
plt.plot(xs, -np.log(1 - xs)), plt.title('-log (-(x-1))'), plt.grid()   # x를 1만큼 평행이동후 y축 대칭, x축 대칭
plt.show()
```

```python
plt.figure(figsize=(6, 10))
plt.plot(xs, -np.log(xs), lw=2)     # y의 실제값이 1일때 사용하는 손실 값
plt.plot(xs, -np.log(1-xs), lw=2)   # y의 실제값이 0일때 사용하는 손실 값
plt.title('Binary Crossentropy')
plt.grid()
plt.show()
```
